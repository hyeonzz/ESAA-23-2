### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv).

```python
data = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv')
data
```

### Step 3. Assign it to a variable called tips

```python
tips = data
```

### Step 4. Delete the Unnamed 0 column

```python
tips_drop=tips.drop('Unnamed: 0',axis=1)
tips_drop
```

### Step 5. Plot the total_bill column histogram

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(tips_drop['total_bill'], bins=10)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips_drop['total_bill'],
    y=tips_drop['tip']
    )
axes1.set_title('Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
fig = sns.pairplot(tips_drop)
```

### Step 8. Present the relationship between days and total_bill value

```python
ax = plt.subplots()
ax = sns.barplot(x='day', y='total_bill', data=tips_drop)
ax.set_title('Bar plot of average total bill for day')
ax.set_xlabel('Day')
ax.set_ylabel('Average total bil')
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
tips_drop['sex_color'] = tips_drop['sex'].apply(recode_sex)
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips_drop['tip'],
    y=tips_drop['day'],
    c=tips_drop['sex_color'],
    )
axes1.set_title('Tip vs Day Colored by Sex')
axes1.set_xlabel('Tip')
axes1.set_ylabel('Day')
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips_drop)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
lunch = tips_drop[tips_drop['time'] == 'Lunch']
dinner = tips_drop[tips_drop['time']== 'Dinner']
```

```python
ax_lunch = plt.subplot(1,2,1)
ax_dinner = plt.subplot(1,2,2)

ax_lunch.hist(lunch['tip'], bins=10)
ax_lunch.set_title('Tip Histogram at Lunch')
ax_lunch.set_xlabel('Tip')
ax_lunch.set_ylabel('Frequency')

ax_dinner .hist(dinner['tip'], bins=10)
ax_dinner.set_title('Tip Histogram at Dinner')
ax_dinner.set_xlabel('Tip')
ax_dinner.set_ylabel('Frequency')

fig
```
